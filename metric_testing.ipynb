```python
from data_prep import prep_vermont
import plotly.figure_factory as ff
import plotly.graph_objs as go
import numpy as np
```

```python
colorscale = ["#f7fbff", "#ebf3fb", "#deebf7", "#d2e3f3", "#c6dbef", "#b3d2e9", "#9ecae1",
    "#85bcdb", "#6baed6", "#57a0ce", "#4292c6", "#3082be", "#2171b5", "#1361a9",
    "#08519c", "#0b4083", "#08306b"]

def gen_choropleth(values: list, title: str, legend_title: str, fips: list, colorscale=colorscale):

    fig = ff.create_choropleth(
        fips=fips, values=values, scope=['VT'],
        binning_endpoints=[], colorscale=colorscale,
        county_outline={'color': 'rgb(255,255,255)', 'width': 0.5}, round_legend_values=True,
        legend_title=legend_title, title=title
    )

    return fig
```

```python
df, counties, fips = prep_vermont()
df = df.reset_index(drop=True)
condensed_counties = df.groupby('County').mean()

mean_poverty = condensed_counties['PovertyRate']
mean_poverty = mean_poverty.round(3).values
mean_pop = condensed_counties['Pop2010']
mean_pop = mean_pop.round(3).values
```

```python
num_tracts = dict(zip(set([i for i in df['County']]), df.groupby('County')['State'].count().values))
l = list(num_tracts.values())
num_tract = []
for item in l:
    num_tract.append(float(item))
```

```python
tracts = gen_choropleth(num_tract, 'Number of Tracts in Each County', '# of Tracts', fips)
tracts.show()
```

```python
poverty_map = gen_choropleth(mean_poverty, 'Mean Poverty', 'Mean Poverty Levels', fips)
poverty_map.show()
```

```python
pop_map = gen_choropleth(mean_pop, 'Mean Pop', 'Mean Population Levels', fips)
pop_map.show()
```

```python
la1and10 = condensed_counties['LA1and10']
la1and10 = la1and10.round(3).values

la1and10_map = gen_choropleth(la1and10, 'Low Access Tracts (1 mile for Urban | 10 mile for rural)', 'Mean Flag for Tracts in County', fips)
la1and10_map.show()
```

```python
LATracts1 = condensed_counties['LATracts1']
LATracts1 = LATracts1.round(3).values

LATracts1_map = gen_choropleth(LATracts1, 'Low Access Tracts (1 Mile Distance)', 'Mean Flag for Tracts in County', fips)
LATracts1_map.show()
```
